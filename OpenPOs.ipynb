```python
import pandas as pd

POs = pd.read_excel("2023.04.18 CLM_SEBN_Open_POs_detail.xlsx", sheet_name="Open POs - detail")
```

```python
POs
```

```python
print(POs.head())
print(POs.describe())
```

```python
POs1 = POs.groupby(['Plant','Material No'])['Rest'].sum().reset_index()
```

```python
POs1
```

```python
POs1 = POs1[~POs1.Plant.str.contains("SEUA")]
```

```python
POs1
```

```python
POs1['Rest'] = POs1['Rest'].map('{:,.0f}'.format)
```

```python
POs1['PlantPOs'] = POs1['Plant'].astype(str) + "(" + POs1['Rest'].astype(str) + ")"
```

```python
POs1.drop(columns=['Plant', 'Rest'], inplace=True)
```

```python
POs1
```

```python
POs_final = POs1.groupby('Material No', as_index=False).agg({'PlantPOs': ', '.join})
```
